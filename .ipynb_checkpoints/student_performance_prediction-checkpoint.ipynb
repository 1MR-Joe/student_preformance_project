# Encoding "a form of data transformation"

### Ordinal values:  these categorical values are in natural order. We can sort or order them either in increasing or decreasing order. 

use replace function to switch to the numerical form

### Nominal values: these are regular categorical values. You can't order nominal values. 

use get_dummies() by pandas

more on these types of values here: https://www.scribbr.com/statistics/nominal-data/

also check that one out:
https://www.dataquest.io/blog/machine-learning-preparing-data/

another one, cuz why not:
https://www.datacamp.com/tutorial/tutorial-data-cleaning-tutorial#How%20Data%20Cleaning%20Is%20Done

```python
# school => school_gp: 1 for GP, 0 for MS
# sex => in_female: 1 for 'female', 0 for 'male'
# Pstatus => are_parents_together: 1 for T 'together', 0 for A 'apart'
# address => is_urban: 1 for U 'urban', 0 for R 'rural'
# famsize => famsize_LE3: 1 for LE3, 0 for GT3

# Medu, Fedu are ok

# Mjob, Fjob, reason => get_dummies

# guardian => guardian_is_a_parent: mother or father = 1, other = 0

# traveltime, studytime, failures are ok

# schoolsup: 1 for yes, 0 for no
# famsup: 1 for yes, 0 for no
# paid: 1 for yes, 0 for no
# activities: 1 for yes, 0 for no
# nursery: 1 for yes, 0 for no
# higher: 1 for yes, 0 for no
# romantic: 1 for yes, 0 for no
# internet: 1 for yes, 0 for no
# the rest are ok
```

### Binary variables

```python
transform = pd.DataFrame()
transform['school_gp'] = df.school.replace({'GP': 1, 'MS': 0})
transform['is_female'] = df.sex.replace({'F': 1, 'M': 0})
transform['urban_address'] = df.address.replace({'U': 1, 'R': 0})
transform['famsize_LE3'] = df.famsize.replace({'LE3': 1, 'GT3': 0})
transform['parents_together'] = df.Pstatus.replace({'T': 1, 'A': 0})
transform['guardian_is_parent'] = df.guardian.replace({'mother': 1, 'father': 1, 'other': 0})
transform['schoolsup'] = df.schoolsup.replace({'yes': 1, 'no': 0})
transform['famsup'] = df.famsup.replace({'yes': 1, 'no': 0})
transform['paid'] = df.paid.replace({'yes': 1, 'no': 0})
transform['activities'] = df.activities.replace({'yes': 1, 'no': 0})
transform['nursery'] = df.nursery.replace({'yes': 1, 'no': 0})
transform['higher'] = df.higher.replace({'yes': 1, 'no': 0})
transform['romantic'] = df.romantic.replace({'yes': 1, 'no': 0})
transform['internet'] = df.internet.replace({'yes': 1, 'no': 0})

ready_cols = ['age', 'Medu', 'Fedu', 'traveltime', 'studytime', 'failures', 'famrel', 'freetime', 'goout', 'Dalc', 'Walc', 'health', 'absences', 'G1', 'G2', 'G3']
transform = pd.concat([transform, df[ready_cols]], axis = 1)
transform.info()
```

```python
# father's job
# mother's job
# reason
nominal_columns = ['Fjob', 'Mjob', 'reason']
dummy_df = pd.get_dummies(df[nominal_columns])
ready_df = pd.concat([transform, dummy_df], axis = 1)
ready_df.info()
```

# making a decision tree to determine the feature importance

```python
from sklearn.tree import DecisionTreeClassifier

# creating the decision tree classifier object
tree = DecisionTreeClassifier()
```

```python
x = ready_df.drop(['G1', 'G2', 'G3'], axis = 1)
y = ready_df[['G1', 'G2', 'G3']]

# training the model
tree.fit(x, y)

# getting the feature importance
importances = tree.feature_importances_

# converting importances to a dataframe
df_importances = pd.DataFrame({
    'Feature': x.columns.to_list(),
    'importance': importances
})
df_importances.sort_values(by = 'importance',ascending = False)
```

# Data splitting & data fitting

```python
from sklearn.model_selection import train_test_split
from sklearn import linear_model

x = ready_df.drop(['G1', 'G2', 'G3'], axis = 1)
y = ready_df['G1']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, shuffle = True)
model = linear_model.LinearRegression()
model.fit(x_train, y_train)

prediction = model.predict(x_test)

plt.plot(np.arange(len(y_test)), y_test)
plt.plot(np.arange(len(y_test)), prediction)
```

```python
plt.figure(figsize = (30, 5))
plt.bar(np.arange(len(y_test)), y_test)
plt.bar(np.arange(len(y_test)), prediction)
plt.show()
# double click the plot
```
