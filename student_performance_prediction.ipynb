# Encoding "a form of data transformation"

### Ordinal values:  these categorical values are in natural order. We can sort or order them either in increasing or decreasing order. 

use replace function to switch to the numerical form

```python
# are_parents_together: 1 for T 'together', 0 for A 'apart'
# is_urban: 1 for U 'urban', 0 for R 'rural'
# famsize_LE3: 1 for LE3, 0 for GT3
# guardian_is_a_parent: mother or father = 1, other = 0
# schoolsup: 1 for yes, 0 for no
# famsup: 1 for yes, 0 for no
# paid: 1 for yes, 0 for no
# nursery: 1 for yes, 0 for no
# romantic: 1 for yes, 0 for no
```

```python
df1.info()
```

```python
df1.guardian
```

```python
# check the data type after transforming
transformation = pd.DataFrame()
transformation['is_urban'] = df1.address == 'urban'
transformation['famsize_LE3'] = df1.famsize == 'LE3'
transformation['are_parents_together'] = df1.Pstatus == 'T'
transformation['guardian_is_a_parent'] = ((df1.guardian == 'father') | (df1.guardian == 'mother'))
transformation['schoolsup'] = df1.schoolsup == 'yes'
transformation['famsup'] = df1.famsup == 'yes'
transformation['paid'] = df1.paid == 'yes'
transformation['nursery'] = df1.nursery == 'yes'
transformation['romantic'] = df1.romantic == 'yes'
transformation['higher'] = df1.higher == 'yes'
transformation.info()
```

```python
df2 = df1.drop(['address', 'famsize', 'Pstatus', 'guardian', 'schoolsup', 'famsup', 'paid', 'nursery', 'romantic', 'higher'], axis = 1)
transformed = pd.concat([df2, transformation], axis = 1)
transformed
```

### Nominal values: these are regular categorical values. You can't order nominal values. 

use get_dummies() by pandas

more on these types of values here: https://www.scribbr.com/statistics/nominal-data/

```python
# father's job
# mother's job
# reason
nominal_columns = ['Fjob', 'Mjob', 'reason']
dummy_df = pd.get_dummies(transformed[nominal_columns])
nominal_trans = pd.concat([transformed, dummy_df], axis = 1)
nominal_trans = nominal_trans.drop(nominal_columns, axis = 1)
nominal_trans.columns.tolist()
```
